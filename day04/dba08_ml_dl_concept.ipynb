### 머신러닝, 딥러닝 기초

- 지능(Intelligence) - 어떠한 문제를 해결하기 위한 지적(뇌를 사용한)인 활동 능력
- 인공지능(Artificial Intelligence) - 기계가 사람의 지능을 모방하는 기술
- 인공지능 개념이 나온건 1950년대이지만 실제로 사용할 수 있게 되기까지 60년 정도 걸림

#### 인공지능의 목표
- 인간이 하는 모든 일을 인공지능을 탑재한 기계가 모두 해내도록 만드는 것

#### 인공지능 종류
- 약 인공지능과 강 인공지능으로 분류
    - 약 인공지능(Weak AI) - 사람이 쉽게 처리할 수 있었지만, 컴퓨터로 하기 어려웠던 일을 수행할 수 있게 하는 인공지능
    - 지각을 가지고 있지 않음. 특정한 업무만 처리하고 거기에 집중함
    - 현재를 약 인공지능의 시대라 칭함

- 강 인공지능(Strong AI)
    - AGI(Artificial General Intelligence) 성능을 가진 인공지능
    - 인공지능 연구의 목표
    - 약 인공지능에 비해 더 인간같고, 물어보면 다 대답해주는 로봇이 예
    - http://youtube.com/watch?v=bAr75K-5Uqw?si=FJY8WVUm3d8NTD7W (참조)
    - 강 인공지능으로 가고 있음

#### 인공지능의 발전 요인
- 인공지능이 60년동안 발전할 수 없었지만 이제 발전하게 된 이유

- 컴퓨터 하드웨어 발전
    - 기타 하드웨어도 포함하지만, CPU보다
    - GPU(Graphic Processing Unit)가 괄목할 발전, NVIDIA/AMD 등의 그래픽카드를 의미
    - 3D 게임/그래픽, 블록체인 계산, 딥러닝 계산처리
    - NPU(Neural network Processing Unit) > TPU(Tensor Processing Unit) 구글에서 만든 NPU

- 기하급수적인 데이터 증가
    - 빅데이터 등으로 정형/비정형 데이터 폭발적 증가
    - 하루에 생산되는 데이터 총량 2.5EB(엑사바이트 / 10억 기가바이트)

- 알고리즘 발전
    - 하드웨어와 데이터 사용해서 처리할 수 있는 알고리즘이 개선
    - 새로운 인공지능 알고리즘도 개발

#### 머신러닝과 딥러닝
- https://blogs.nvidia.co.kr/blog/difference_ai_learning_machinelearning/  발췌

<img src="https://blogs.nvidia.co.kr/wp-content/uploads/sites/16/2016/08/postfiles8.naver_.png"  width="800">

- 머신러닝(Machine Learning)
    - 데이터 학습 기반의 인공지능 분야
    - 컴퓨터에게 어떻게 동작할지 코드로 명시하지 않고, 데이터를 이용해서 학습할 수 있도록 알고리즘과 기술 개발
    - 전문가 시스템이라고도 호칭

- 딥러닝(Deep Learning)
    - 인공신경망 알고리즘을 기반한 비정형데이터 분석에 뛰어난 성능을 발휘하는 인공지능 분야
    - 텍스트, 영상, 음성 등에 장점, DB등의 정형 데이터도 분석 뛰어남

- 프로세스 차이 - 일반적 프로그래밍과 딥러닝/머신러닝 차이

    

#### 머신러닝/딥러닝으로 할 수 있는 일
- 머신러닝 - 회귀, 분류, 군집, 텍스트 마이닝 
- 딥러닝 - 머신러닝 + 컴퓨터 비전, 패턴인식, 물체인식, 실시간 행동분석, 새로운 이미지 만들기, 자율주행, 로봇

#### 머신러닝 분류
- 지도학습 - 훈련데이터로 하나의 함수를 유추해내는 방법 > 회귀, 분류
- 비지도학습 - 레이블 없는 훈련데이터를 이용해서 스스로 학습을 하는 방법 > 군집, 차원축소, 이상치 탐지, 
- 강화학습 - 학습 시스템을 에이전트라 부르고, 에이전트가 취한 행동에 따라 보상/벌점을 주어 가장 큰 보상을 받는 방향으로 유도하는 학습

#### 딥러닝은
- 뉴럴 네트워크를 사용, 머신러닝보다 정확도 높이는 것에 집중

### 머신러닝, 딥러닝 파이썬 라이브러리

#### Pandas
- 데이터 과학분야 사용되는 오픈소스 라이브러리. 데이터 조작, 정리 등의 핵심도구

#### NumPy
- 과학 연산에 초점을 맞춘 오픈소스 라이브러리. 행렬 계산에 특화, 난수 생성 등

#### SciPy
- Numpy 확장 라이브러리. 과학컴퓨팅, 데이터처리, 고성능 컴퓨팅 오픈소스 라이브러리

#### Scikit Learn
- 파이썬의 대표적인 머신러닝 라이브러리(딥러닝 없음)
- 지도 및 비지도 학습 알고리즘 개발, 적용
- NumPy, SciPy, Matplotlib 패키지와 통합되어 머신러닝 프로젝트에 필수적으로 사용
- 회귀, 분류, 차원축소, 클러스터링(군집) 등 기능 포함 가장 최초에 머신러닝을 배울 때 많이 사용
- 쉽게 배울 수 있음!!

#### Theano
- 수학계산, 식정의, 최적화 라이브러리 중 하나. 대규모 다차원 배열 처리
- NumPy와 통합, GPU 사용가능, 미분이 쉽고, 속도와 안정성 최적화, C코드 생성
- 딥러닝 연구 및 개발에 업계 표준으로 사용된다??

#### Tensorflow
- 구글에서 만든 고성능 수치계산 라이브러리. 머신러닝, 딥러닝 알고리즘 포함
- 훈련, 개발하는데 널리 사용되는 오픈소스 라이브러리
- 2015년 발표, 파이썬 C++, CUDA로 개발
- 딥러닝 분야를 거의 장악했었음!
- 구글 코랩에서 바로 사용
- 2.0으로 버전업 되면서 Keras를 포함함. 1.x와 2.x는 상이함

#### PyTorch
- 메타 AI(페이스북)에서 만든 머신러닝, 딥러닝 라이브러리
- 2016년 10월 발표, 파이썬, C++, CUDA로 개발
- 텐서플로우 비직관적인 구조와 난이도 때문에 사용자가 늘고있는 추세
- 연구개발에는 파이토치가 더 많이 사용되고 있음
- 2022년도 기준으로 텐서플로 사용률 넘어섬
- 영상처리 분야 Yolo 라이브러리가 PyTorch 기반임
